```python
# Print unique values in the 'job_title_sim' column before cleaning
print("Before cleaning:")
print(df['job_title_sim'].unique())

# Clean the 'job_title_sim' column
df['job_title_sim'] = df['job_title_sim'].str.strip()  # Remove leading/trailing whitespace
df['job_title_sim'] = df['job_title_sim'].str.lower()  # Convert to lowercase
df['job_title_sim'] = df['job_title_sim'].replace({
    'data scientist': 'Data_Scientist',
    'other scientist': 'Other_Scientist',
    'analyst': 'Data_Analyst',
    'data analitics': 'Data_Analyst',
    'na': 'Other',
    'data modeler': 'Data_Modeler',
    'data scientist project manager': 'Project_Manager',
    'machine learning engineer': 'MachineLearning_Engineer',
    'director': 'Director',
    'data engineer':'Data_Engineer'
})

# Rename the 'job_title_sim' column to 'Occupation'
df.rename(columns={'job_title_sim': 'Occupation'}, inplace=True)

# Print unique values in the 'Occupation' column after cleaning
print("\nAfter cleaning:")
print(df['Occupation'].unique())
```

```python
# Define the skills columns
skills_columns = ['Python', 'spark', 'aws', 'excel', 'sql', 'sas', 'keras', 'pytorch', 'scikit', 'tensor', 'hadoop', 'tableau', 'bi', 'flink', 'mongo', 'google_an']

# Function to get required skills
def get_required_skills(row):
    skills = []
    for skill in skills_columns:
        if row[skill] == 1:
            skills.append(skill)
    return ', '.join(skills)

# Apply the function to each row of the DataFrame
df['RequiredSkills'] = df.apply(get_required_skills, axis=1)
```
